```python
import matplotlib.pyplot as plt # Única lib externa permitida

# Definir a classe Livro (molde)
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        # atributos do livro
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

#  lista de livros que guardara os livros (objetos)
lista_de_livros = []

# função para testar o sistema + dados
def carregar_dados_teste():
    lista_de_livros.append(Livro("O Senhor dos Aneis: A Sociedade do Anel", "J.R.R. Tolkien", "Fantasia epica", 3))
    lista_de_livros.append(Livro("Harry Potter e a Pedra Filosofal", "J.K. Rowling", "Fantasia juvenil", 5))
    lista_de_livros.append(Livro("Harry Potter e a Camara Secreta", "J.K. Rowling", "Fantasia juvenil", 2))
    lista_de_livros.append(Livro("A Guerra dos Tronos", "George R.R. Martin", "Fantasia epica", 4))
    lista_de_livros.append(Livro("1984", "George Orwell", "Distopia", 1))
    lista_de_livros.append(Livro("Dom Casmurro", "Machado de Assis", "Romance", 2))

carregar_dados_teste()

# Cadastrs os atributos/trata erros de entrada/pega os atributos, tranforma em um objeto, coloca
# na variável novo_livro e adiciona na lista lista_de_livros
def cadastrar_livro():
    print("\n--- Novo Cadastro ---")
    titulo = input("Digite o titulo do livro: ")
    autor = input("Digite o autor do livro: ")
    genero = input("Digite o genero do livro: ")
    while True:
      try:
          quantidade = int(input("Digite a quantidade disponivel do livro: "))
          if quantidade < 0:
              print("A quantidade nao pode ser negativa. Digite um numero inteiro positivo.")
              continue
          break
      except ValueError:
          print("Entrada invalida. Digite um numero inteiro para a quantidade.")
    novo_livro = Livro(titulo, autor, genero, quantidade)
    lista_de_livros.append(novo_livro)
    print(f"Livro '{titulo}' cadastrado com sucesso!")

# verifica a existência de livros/mostra todos os livros caso tenha livros
def listar_livros():
    print("\n--- Lista de Livros Cadastrados ---")
    if len(lista_de_livros) == 0:
        print("Nenhum livro cadastrado ainda.")
    else:
        for i, livro in enumerate(lista_de_livros, 1):
            print(f"{i}. Titulo: {livro.titulo} | Autor: {livro.autor} | Genero: {livro.genero} | Quantidade: {livro.quantidade}")

# hora de buscar os livros pelo título
def buscar_livro():
    print("\n--- Buscar Livro pelo Titulo ---")
    termo_original = input("Digite o titulo do livro: ").strip()# formatação para espaços
    termo_busca = termo_original.lower() # formatação para A, a

    resultados = []
    # apartir da entrada é verificado se o títlo está na lista mãe, se sim, adiciona tudo na lsita resultados
    for livro in lista_de_livros:
        if termo_busca in livro.titulo.lower():
            resultados.append(livro)
    # mostra a quantidade/livros se caso achar algum livro na lista resultados
    print(f"\n--- Resultados para '{termo_original}' ---")
    if len(resultados) == 0:
        print("Nenhum livro encontrado com esse titulo.")
    else:
        print(f"Encontrados {len(resultados)} livro(s):")
        for i, livro in enumerate(resultados, 1):
            print(f"{i}. Titulo: {livro.titulo} | Autor: {livro.autor} | Genero: {livro.genero} | Quantidade: {livro.quantidade}")

# Utiliza Matplotlib para gerar gráficos
def gerar_grafico():
    print("\n--- Gerando Grafico ---")
    if len(lista_de_livros) == 0:
        print("A estante esta vazia! Cadastre livros primeiro.")
        return
    contagem_generos = {}
    for livro in lista_de_livros:
        contagem_generos[livro.genero] = contagem_generos.get(livro.genero, 0) + livro.quantidade
    # para poder gerar o gráfico, tem que ter duas listas ("""list()"""")
    # esse comando pega o objeto e desempacota em uma lsita novamente
    # o Matplotlib requer duas listas para seus eixos X, Y
    nomes = list(contagem_generos.keys())
    quantidades = list(contagem_generos.values())
    # partes do gráfico
    plt.bar(nomes, quantidades)
    plt.title('Quantidade de Livros por Genero')
    plt.xlabel('Generos')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show() # mostra o gráfico
    plt.close() # sem isso, o programa fica parado sem as opções
print("*** SISTEMA DE GERENCIAMENTO DE LIVROS ***")

# menu de controle do sistema
while True:
    print("\n--- MENU PRINCIPAL ---")
    print("1 - Cadastrar novo livro")
    print("2 - Listar todos os livros")
    print("3 - Buscar um livro pelo titulo")
    print("4 - Gerar grafico")
    print("5 - Sair do sistema")
    opcao = input("Escolha uma opcao (1/2/3/4/5): ")

    if opcao == '1':
        cadastrar_livro()
    elif opcao == '2':
        listar_livros()
    elif opcao == '3':
        buscar_livro()
    elif opcao == '4':
        gerar_grafico()
    elif opcao == '5':
        print("Encerrando o sistema. Ate logo!")
        break
    else:
        print("Opcao invalida! Tente novamente.")
```
