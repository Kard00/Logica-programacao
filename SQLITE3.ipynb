```python
# cria uma tabela
creat_table = """
CREATE TABLE IF NOT EXISTS Produtos
(
  id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
  nome TEXT NOT NULL,
  preco REAL NOT NULL,
  estoque INTEGER
);"""
cursor.execute(creat_table); # executa a criação da tabela
```

```python
conn.commit() # salva todas as laterações
```

```python
conn.close() # encerra o banco de dados
```

```python
import sqlite3

conn = sqlite3.connect(' exemplo.db')
cursor = conn.cursor()

# dados do novo produto
novo_produto = ('Camiseta', 19.99, 50)

# Comando SQL para inserir o novo produto na tabela
inserir_produto = "INSERT INTO Produtos (nome, preco, estoque) VALUES (?, ?, ?)"

# Executando o comando SQL para inserção
cursor.execute(inserir_produto, novo_produto)

# Confirmando as alterações no banco de dados
conn.commit()

# Fechando a conexão
conn.close()
```

```python
import sqlite3

# 1. Conecta ao banco de dados novamente
conn = sqlite3.connect(' exemplo.db')
cursor = conn.cursor()

# 2. Comando SQL para buscar (selecionar) tudo da tabela Produtos
cursor.execute("SELECT * FROM Produtos")

# 3. Pega todos os resultados que o banco encontrou e guarda numa variável
produtos = cursor.fetchall()

# 4. Mostra os produtos na tela
print("Produtos no banco de dados:")
for produto in produtos:
    print(produto)

# 5. Sempre fechar a conexão no final
conn.close()
```
