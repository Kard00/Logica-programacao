```python
# 1. Importa as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 2. Carrega o CSV. Sep=';' porque planilha BR usa ponto-e-vírgula
df = pd.read_csv('consumo-de-energia-eletrica-no-palacio-do-jaburu-2017-2024-setembro.csv',
                 encoding='latin-1', sep=';')

# 3. Limpa nome das colunas - tira espaços sobrando
df.columns = df.columns.str.strip()

# 4. Transforma 'VALOR BRUTO DA FATURA' em número
# Porque ele vem como texto: "R$ 17.910,77"
def limpar_valor(valor):
    valor = valor.replace('R$', '').replace(' ', '').replace('.', '').replace(',', '.')
    return float(valor)

df['VALOR_NUM'] = df['VALOR BRUTO DA FATURA'].apply(limpar_valor)
```
